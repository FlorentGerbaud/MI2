## Exercice 1 (5 points)
On se place sur l'espace $\ell^2$ des suites réelles de carré sommable muni de son produit scalaire usuel :
si $X=(x_n)_n$ et $Y=(y_n)_n$ on a

$$ (X|Y) = \sum_{n \geq 0} x_n y_n. $$

### 1.1
Soit $k \geq 1$. Montrer que la suite $X_k := (1,\dots,1,0,\dots)$ dont seuls les $k$ premiers termes sont non nuls et égaux à $1$ est dans $\ell^2$.

**Réponse.** La suite est presque nulle, donc en particulier appartient à $\ell^2$.

### 1.2
On note $F := \{X=(x_n)_n \in \ell^2\ |\ x_1+x_3=0\}$. Montrer que $F$ est un sous-espace vectoriel fermé de $\ell^2$.

**Réponse.** La partie $F$ est un sev fermé comme orthogonal de $\{Y\}$ avec
$Y=(0,1,0,1,0,\dots)$.

### 1.3
Déterminer le projeté orthogonal de $X_k$ sur $F$. 

**Réponse.** Le projeté $\bar{X}_k$ est caractérisé par le fait que $X_k-\bar{X}_k$ appartient à $F^\perp=\{Y\}^{\perp\perp}=\mathrm{Vect}(\{Y\})$ : il existe donc un réel $\lambda$ tel que
$X_k-\bar{X}_k = \lambda Y$. Comme $\bar{X}_k$ appartient à $F$, on a $(\bar{X}_k|Y)=0$, d'où

$$ \lambda = \frac{(X_k|Y)}{\|Y\|^2} \cdot $$

On vérifie $\lambda=0$ si $k=1$, $\lambda=1/2$ si $k=2,3$, et $\lambda=1$ si $k \geq 4$.

### 1.4
En déduire la distance de $X_k$ à $F$.

**Réponse.** $d(X_k,F)=\|X_k-\bar{X}_k\|=|\lambda|\,\|Y\|$ qui vaut $0$ si $k=1$,
$\frac{\sqrt{2}}{2}$ si $k=2,3$, et $\sqrt{2}$ si $k \geq 4$.

## Exercice 2 (5 points)
Sur $L^2([0,\pi])$ muni de son produit scalaire usuel, on note $F$ le sous-espace vectoriel engendré par la fonction $\cos$.

### 2.1
Montrer que $F$ est un sous-espace vectoriel fermé de $L^2([0,\pi])$.

**Réponse.** C'est un sev de dimension finie (dimension un), donc un sev fermé.

### 2.2
Montrer que la fonction $\exp$ appartient à $L^2([0,\pi])$ et déterminer son projeté orthogonal sur $F$.

**Réponse.** Le projeté orthogonal $\bar{f}$ est caractérisé par la relation
$(\exp-\bar{f}|\cos)=0$, sachant qu'il existe un réel $\lambda$ tel que $\bar{f}=\lambda \cos$. On en déduit $\lambda=(\exp|\cos)/\|\cos\|^2$ avec

$$ (\exp|\cos) = \mathrm{Re} \int_0^\pi e^t e^{it}\,\mathrm{d} t = -\frac{e^\pi+1}{2},\quad
   \|\cos\|^2 = \frac{\pi}{2}\,, $$

d'où $\bar{f} =-\frac{e^\pi+1}{\pi}\cos$.

### 2.3
Orthonormaliser la famille $\{\cos,\exp\}$ dans $L^2([0,\pi])$.

**Réponse.** Par Gram-Schmidt, on pose $e_0=\sqrt{2/\pi}\,\cos$ et
$e_1=\pm(\exp-\bar{f})/\|\exp-\bar{f}\|$ avec

$$ \|\exp-\bar{f}\|^2 = \frac{1}{2}(e^{2\pi}-1)-\frac{1}{2\pi}(e^\pi+1)^2. $$
