## Exercice 1 (3 points)
Montrer que l'application $f : \mathbf{R}^2 \to \mathbf{R}^4$ définie par

$$ f(x_1,x_2) := \left[ \begin{array}{c}
  x_1\cos x_2+x_2\sin x_1\\
  x_2\sin(x_1-x_2^2)\\
  x_1-x_2\\
  2x_2^3
 \end{array} \right] $$

est dérivable et donner l'expression de sa dérivée.

**Réponse.**

$$ f'(x_1,x_2) = \left[ \begin{array}{cc}
  \cos x_2+x_2\cos x_1 & -x_1\sin x_2+\sin x_1\\
  x_2\cos(x_1-x_2^2) & \sin(x_1-x_2^2)-2x_2^2\cos(x_1-x_2^2)\\
  1 & -1\\
  0 & 6x_2^2 \end{array} \right] $$

## Exercice 2 (4 points)
### 2.1
Montrer que l'application $f : \mathbf{R}^n \to \mathbf{R}$ définie par

$$ f(x) := \frac{1}{1+\|x\|^2} $$

est dérivable et donner l'expression de son gradient. La norme désigne la norme
euclidienne sur $\mathbf{R}^n$,

$$ \|x\|=(|x_1|^2+\cdots+|x_n|^2)^{1/2}. $$

**Réponse.** L'application est dérivable comme composée $x \mapsto \|x\|^2 \mapsto 1/(1+\|x\|^2)$ et

$$ \nabla f(x) =-\frac{2x}{(1+\|x\|^2)^2} \cdot $$

### 2.2
Montrer que $\nabla f$ est également dérivable et donner
l'expression du hessien de $f$.

**Réponse.** Le gradient est dérivable comme produit des fonctions dérivables $x \mapsto 2x$ et $x \mapsto -1/(1+\|x\|^2)^2$ (la seconde étant à nouveau dérivable comme composée), et

$$ \nabla^2 f(x) = -\frac{2I}{(1+\|x\|^2)^2} + \frac{8x\!\cdot^t\!x}{(1+\|x\|^2)^3} \cdot $$

## Exercice 3 (6 points)
### 3.1
Mettre l'équation différentielle

$$ \dot{x}(t)=x^2(t)-1 $$

sous la forme $\dot{x}(t)=f(t,x(t))$ avec $f:\mathbf{R}^2 \to \mathbf{R}$ que l'on précisera.

**Réponse.**

$$ f(t,x) = x^2-1 $$

### 3.2
Étant donné $(t_0,x_0) \in \mathbf{R}^2$, justifier que l'équation différentielle

$$ \dot{x}(t)=x^2(t)-1,\quad x(t_0)=x_0, $$

possède une unique solution maximale.

**Réponse.** La fonction $f$ est de classe $\mathscr{C}^1$, donc le théorème de Cauchy-Lipschitz s'applique.

### 3.3
Déterminer la solution maximale de l'équation différentielle

$$ \dot{x}(t)=x^2(t)-1,\quad x(0)=x_0 $$

pour $x_0=1$, puis pour $x_0=-1$.

**Réponse.** Clairement $x(t)=1$, $t \in \mathbf{R}$ est solution dans le premier cas (définie sur tout $\mathbf{R}$ donc maximale), et $x(t)=-1$, $t \in \mathbf{R}$ dans le second.

### 3.4
On considère l'équation différentielle

$$ \dot{x}(t)=x^2(t)-1,\quad x(0)=0. $$

Justifier que la solution maximale est à valeurs dans $]-1,1[$,
puis déterminer cette solution maximale.

**Réponse.** Comme $x(0)=0 \neq \pm 1$, le graphe de la solution maximale ne coupe pas les graphes des deux solutions constantes de la question précédente (unicité) : par continuité, cette solution maximale est nécessairement à valeurs dans $]-1,1[$. On peut donc écrire

$$ \frac{\dot{x}(t)}{x^2(t)-1} = 1 $$

pour séparer les variables, et intégrer sur $[0,t]$. Comme $1/(x^2-1) = (1/2)(1/(x-1)-1/(x+1))$, on déduit

$$ \ln\frac{|x(t)-1|}{|x(t)+1|}
 = \ln\frac{1-x(t)}{1+x(t)} = 2t, $$
 
d'où $x(t)=2/(e^{2t}+1)-1$, $t \in \mathbf{R}$ (globale, donc maximale).

## Exercice 4 (7 points)
On munit l'espace vectoriel $E$ des applications bornées ($*$) de $[0,1]$ dans $\mathbf{R}$ de la norme

$$ \|x\|_\infty := \sup_{t \in [0,1]} |x(t)|. $$

($*$) Une application $x : [0,1] \to \mathbf{R}$ est bornée s'il existe un réel $M$ tel que, pour tout $t$ dans $[0,1]$, $|x(t)| \leq M$.

### 4.1
Montrer qu'on définit bien ainsi une norme.

**Réponse.** Si l'application $x$ est bornée, l'ensemble $\{|x(t)|,\ t \in [0,1]\}$ est non vide et majoré, donc possède bien une borne supérieure. Le caractère défini positif et l'homogénéité positive sont évidents et, si $x$ et $y$ sont dans E, pour $t \in [0,1]$ on a

$$ |x(t)+y(t)| \leq |x(t)| + |y(t)| \leq \sup_{t \in [0,1]} |x(t)| + \sup_{t \in [0,1]} |y(t)|, $$

de sorte que la borne supérieure du membre de gauche est plus petite que le majorant à droite ; ce qui montre l'inégalité triangulaire.

### 4.2
Soit $(x_n)_n \in E^\mathbf{N}$ une suite de Cauchy de cet espace vectoriel normé. Soit $t \in [0,1]$, montrer que la suite $(x_n(t))_n \in \mathbf{R}^\mathbf{N}$ est également une suite de Cauchy. Que peut-on en déduire ?

**Réponse.** Soit $\varepsilon > 0$ ; la suite étant de Cauchy, il existe un rang $N$ tel que, pour tous $p$ et $q$ supérieurs à $N$, $\|x_p-x_q\|_\infty \leq \varepsilon$. En particulier, si $t \in [0,1]$, on a $|x_p(t)-x_q(t)| \leq \varepsilon$, ce qui montre que $(x_n(t))_n$ est une suite de Cauchy sur $(\mathbf{R},|.|)$. Ce dernier espace étant complet, cette suite converge.

### 4.3
Soit $(x_n)_n \in E^\mathbf{N}$ une suite qui converge dans $(E,\|\cdot\|_\infty)$ vers une fonction $x$. On suppose de plus chaque $x_n$ continue de $[0,1]$ dans $\mathbf{R}$. Soit $\varepsilon \gt 0$ ; montrer qu'il existe un rang $n \in \mathbf{N}$ tel que, pour tout $t$ dans $[0,1]$, $|x(t)-x_n(t)| \leq \varepsilon$.

**Réponse.** Conséquence immédiate de la convergence de la suite. (Il existe même un rang *à partir duquel* cette inégalité est vraie.)

### 4.4
Soit $t_0$ dans $[0,1]$, en déduire que $x$ est continue en $t_0$.

**Réponse.** Soit $\varepsilon > 0$ ; comme $x_n$ (rang $n$ de la question précédente) est continue en $t_0$, il existe $\eta > 0$ tel que, si $|t-t_0| \leq \eta$, on a $|x_n(t)-x_n(t_0)| \leq \varepsilon$. Pour de tels $t$ on a donc

$$ |x(t)-x(t_0)| \leq |x(t)-x_n(t)|+|x_n(t)-x_n(t_0)|+|x_n(t_0)-x_n(t_0)| \leq 3\varepsilon, $$

d'où la continuité en $t_0$ (et donc en tout point de $[0,1]$).

### 4.5
Que peut-on conclure ?

**Réponse.** Que le sev $\mathscr{C}^0([0,1],\mathbf{R})$ de $E$ (une application continue sur un compact est bornée) est fermé.
