## Exercice 3 (5 points)
On cherche les distributions $S$ dans $\mathscr{D}'(\mathbf{R})$ qui vérifient

$$ x\cdot S'=T_1, \tag{1} $$

où $T_1$ désigne la distribution régulière associée à la fonction constante égale à $1$.

### 3.1
Montrer que $x \cdot \mathrm{vp}(1/x)=T_1$. On rappelle que la distribution valeur principale de $1/x$ est définie par

$$ \langle \mathrm{vp}\frac{1}{x},\varphi \rangle =
   \lim_{\varepsilon \to 0+} \int_{|x| \geq \varepsilon} \frac{\varphi(x)}{x}\,\mathrm{d}x,
   \quad \varphi \in \mathscr{D}(\mathbf{R}). $$
   
### 3.2
En déduire que $S$ est solution de (1) si et seulement si 

$$ x \cdot (S-T_{\ln|x|})'=0. $$

### 3.3
En déduire que $S$ est solution de (1) si et seulement s'il existe un réel $c$ tel que

$$ (S-T_{\ln|x|}-cT_H)' = 0 $$

où $H$ est la fonction de Heaviside.
 
### 3.4
En déduire l'ensemble des solutions de (1).
 
## Exercice 4 (5 points)
On considère le problème suivant : trouver $u \in \mathscr{C}^2([0,1])$ telle que

$$ -u''(x) = f(x),\quad x \in ]0,1[, $$

$$ u(0)=0,\quad u(1)=0, $$

où $f$ est une fonction donnée de $\mathscr{C}^0([0,1])$.

### 4.1
On admet que

$$ (u|v)_{H^1_0} := \int_0^1 u'(x)v'(x)\,\mathrm{d}x $$

définit un produit scalaire qui fait de 

$$ H^1_0 = \{ u \in H^1(]0,1[)\ |\ u(0)=u(1)=0 \} $$

un espace de Hilbert. Montrer que toute solution ("forte") $u$ du problème
est également solution ("faible") du problème suivant : trouver $u$ dans $H^1_0$ tel que, quel que soit $v \in H^1_0$,

$$ (u|v)_{H^1_0} = \int_0^1 f(x)v(x)\,\mathrm{d}x. $$
 
### 4.2
On admet l'existence et l'unicité d'une solution faible.
Montrer que cette solution faible vérifie, au sens des distributions,
l'équation suivante dans $\mathscr{D}'(]0,1[)$ :

$$ -(T_u)''= T_f. $$
 
### 4.3
On déduit de la question précédente que la solution est de
classe $\mathscr{C}^2$ sur $[0,1]$. Montrer finalement que cette solution faible est aussi solution forte.
